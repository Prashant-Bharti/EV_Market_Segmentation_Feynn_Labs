```python
import matplotlib.pyplot as plt
import pandas as pd

# Data

# Create DataFrame
df = pd.read_excel("sales_years.xlsx")

# Plotting
plt.figure(figsize=(12, 6))

# Plot each column with different colors
plt.plot(df['YEAR'], df['2 W'], label='2 W', color='blue', marker='o')
plt.plot(df['YEAR'], df['3 W'], label='3 W', color='green', marker='s')
plt.plot(df['YEAR'], df['4 W'], label='4 W', color='red', marker='^')
plt.plot(df['YEAR'], df['BUS'], label='BUS', color='purple', marker='D')
plt.plot(df['YEAR'], df['TOTAL'], label='TOTAL', color='orange', marker='*')

# Customize the plot
plt.title('Monthly Vehicle Data (Apr-17 to May-18)', fontsize=14)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show plot
plt.show()
```
